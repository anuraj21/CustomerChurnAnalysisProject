```python
def conv (value):
    if value ==1:
        return "yes"
    else:
        return"no"
df['SeniorCitizen']= df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize = (3,4))
ax=sns.countplot( x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("count of Customer by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels= gb.index, autopct = "%1.2f%%")
plt.title("Percentage of churned Customer")
plt.show()
```

```python
#from give pie chart we can conclude that 26.54% of our customer have churned out now let's  explore reason behind it
```

```python


# Assuming df is your DataFrame with the relevant columns
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set the size of the plot grid based on the number of columns
n_rows = 3  # You can adjust this depending on how you want to arrange the plots
n_cols = 3  # Also adjust based on the number of columns

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 15))
axes = axes.flatten()  # Flatten the axes for easy iteration

# Iterate over the columns and axes to create count plots
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i], hue= df["Churn"])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels if needed

# Adjust layout
plt.tight_layout()
plt.show()
```


The image presents multiple count plots showing the distribution of customer churn (Yes/No) across various service categories, such as Phone Service, Internet Service, Online Security, Online Backup, and Streaming Services. In most cases, customers without these additional services tend to have a higher churn rate compared to those who have them. This suggests that customers who subscribe to extra services are more likely to stay with the provider. The data highlights the impact of service offerings on customer retention
